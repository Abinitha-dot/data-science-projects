```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create Retail Sales Dataset
data = {
    'Product': ['Laptop', 'Phone', 'Tablet', 'Headphones', 'Smartwatch'],
        'Sales': [50000, 40000, 15000, 10000, 12000],
            'Profit': [12000, 10000, 4000, 2500, 3000]
            }

df = pd.DataFrame(data)

            # Show dataset
print(df)

            # Sales Bar Chart
sns.barplot(x='Product', y='Sales', data=df)

plt.title("Product Sales Analysis")

plt.show()

            # Profit Bar Chart
sns.barplot(x='Product', y='Profit', data=df)

plt.title("Product Profit Analysis")

plt.show()

            # Statistical Summary
print(df.describe())

            # Conclusion
print("Laptops generated the highest sales and profit.")
```
